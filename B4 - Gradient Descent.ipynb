```python
# Function :  y = (x+3)^2

curr_x = 2          # Initial value of 'x'
rate = 0.01         # Learning rate (determines how much to move)
precision = 0.0001  # Determines when to stop the algorithm (if answer is within precision, then stop)
prev_step_size = 1  #

max_iters = 10000   # Max. number of iterations
iters = 0           # Counter for iterations
```

```python
# Gradient(derivative) of function : y' = 2(x+3)
def gradient(x):
    return 2*(x+3)
```

```python
X = []
while prev_step_size > precision and iters < max_iters:
    prev_x = curr_x                           # Store current x value in prev_x
    curr_x = curr_x - rate * gradient(curr_x)  # Gradient descent
    prev_step_size = abs(curr_x - prev_x)     # Calculate change in x
    iters = iters + 1                        # Increment iteration count
    print("Iteration-"+str(iters)+" :  x = "+str(curr_x))
    X.append(curr_x)
    
print("\nLocal minimum occurs at x="+str(curr_x))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3.5,2.5,1000)
y = (x+3)**2

fig = plt.figure(figsize = (10,5))
plt.plot(x,y)

for x_pt in X: 
    y_pt = (x_pt+3)**2
    plt.plot(x_pt, y_pt,'o')
plt.show()
```
